### 6.1.5. Study Hours

This is numerical variable

```python
minimum  = data_2['study_hours'].min()
maximum = data_2['study_hours'].max()
average = round(data_2['study_hours'].mean(),2)
median = data_2['study_hours'].median()
mode = data_2['study_hours'].mode()
stddev = round(data_2['study_hours'].std(), 3)
var = round(data_2['study_hours'].var(),2)
cv = round(stddev/average, 2)
Q1 = data_2['study_hours'].quantile(0.25)
Q3 = data_2['study_hours'].quantile(0.75)
IQR = Q3-Q1
upper_bound = round(Q3 + 1.5*IQR, 2)
lower_bound = round(Q1 - 1.5*IQR, 2)
number_upper_outliers = data_2[data_2['study_hours'] > upper_bound].shape[0]
number_lower_outliers = data_2[data_2['study_hours'] < lower_bound].shape[0]
kurt = round(data_2['study_hours'].kurt(),2)
skew = round(data_2['study_hours'].skew(),2)

print('minimum value of study hours is: ', minimum)
print('maximum value of study hours is: ', maximum)
print('average value of study hours is: ', average)
print('median value of study hours is: ', median)
print('mode value of study hours is: ', mode)
print('stddev value of study hours is: ', stddev)
print('variance value of study hours is: ', var)
print('coeff of variation value of study hours is: ', cv)

if cv <= 0.10:
    print('study hours is consistent')
elif 0.10 < cv <= 0.30:
    print('study hours has moderate variation')
elif 0.30 < cv <= 0.50:
    print('study hours has high variation')
else:
    print('study hours has extreme variation')

print('first quartile value of study hours is: ', Q1)
print('third quartile value of study hours is: ', Q3)
print('inter quartile range value of study hours is: ', IQR)
print('upper_bound value of study hours is: ', upper_bound)
print('lower_bound value of study hours is: ', lower_bound)
print('number of upper outliers of study hours is: ', number_upper_outliers)
print('number of lower outliers of study hours is: ', number_lower_outliers)
print('kurt value of study hours is ', kurt)

if kurt > 0:
    print('study hours has heavy tails, check for potential outliers')
elif kurt < 0:
    print('study hours has light tails')
else:
    print('study hours is approximately normally distributed')



print('skew value of study hours is ', skew)

if skew > 1:
    print('study hours is highly right skewed')
elif skew < -1:
    print('study hours is highly left skewed')
else:
    print('study hours is normally distibuted')
```

```python
plt.boxplot(data_2['study_hours'], vert = False)
plt.xlabel('study hours')
plt.title('boxplot for study hours')
plt.show()
```

```python
plt.figure(figsize=(7,4))

plt.hist(data_2['study_hours'], bins=16)

plt.xlabel('Study Hours')
plt.ylabel('Number of Students')
plt.title('Distribution of Study Hours')
plt.ylim(600, 1100)
plt.grid(axis='y', alpha=0.3)

plt.show()
```

Study hours range from 0.5 to 8 hours with a mean of 4.26 and median of 4.3, indicating that the typical student studies approximately four hours per day. The coefficient of variation of 0.51 reflects high natural behavioral variability, which is expected for human study patterns rather than data quality issues. Quartile values (Q1 = 2.4, Q3 = 6.1) and an IQR of 3.7 show no evidence of extreme outliers. The distribution is symmetric with near-zero skewness and light-tailed behavior. Study hours therefore represent a stable behavioral feature.
